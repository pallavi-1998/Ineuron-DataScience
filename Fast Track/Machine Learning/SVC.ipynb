## **support vector classifier**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification,make_blobs,make_moons
from sklearn.model_selection import train_test_split
```

```python
X1,y1 = make_classification(n_samples=1000, n_features=2,n_redundant=0, n_informative=2, n_clusters_per_class=1,class_sep=2.0,random_state=42)
```

```python
plt.scatter(X1[:,0], X1[:,1],  c=y1)
plt.show()
```

```python
X_train, y_train, X_test,y_test = train_test_split(X1,y1,test_size=0.2,random_state=42)
```

```python
from sklearn.svm import LinearSVC
clf = LinearSVC()
clf.fit(X_train,y_train)
```

```python
plt.scatter(X2[:,0], X2[:,1],  c=y2)
plt.show()
```
