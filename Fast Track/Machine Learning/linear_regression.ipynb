###  **divide the dataset into dependent and independent feature**

```python
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

###  **divide the dataset into train and test**

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.33,random_state=32)
```

```python
X_train.head()
```

```python
X.shape, X_train.shape, X_test.shape
```

### **feature scaling**  ( perform z-score (x - mean) / std)

```python
from sklearn.preprocessing import StandardScaler  
scaler = StandardScaler()
```

```python
# two steps------------->
# .fit compute std, mean
# .transform compute z-score
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

### **model training**

```python
from sklearn.linear_model import LinearRegression
regression = LinearRegression()
```

### **model fit**

```python
regression.fit(X_train,y_train)
```

### **model coefficent**

```python
regression.coef_
```

### **model intercept**

```python
regression.intercept_
```

### **model prediction**

```python
##v prediction
y_pred = regression.predict(X_test)
```

### **model mse, mae, rmse**

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np
mse = mean_squared_error(y_test,y_pred)
print(mse)
mae = mean_absolute_error(y_test, y_pred)
print(mae)
rmse = np.sqrt(mse)
print(rmse)
```

### **Accuracy r2 and adjusted r sq**

```python
# r square
from sklearn.metrics import r2_score
score = r2_score(y_test,y_pred)
```

```python
#adjusted r square
1 - (1- score)*(len(y)-1)/ (len(y)-X.shape[1]-1)
```

### **Ridge regression**

```python
from sklearn.linear_model import Ridge
ridge = Ridge()
```

```python
ridge = Ridge(alpha=20.0)
ridge.fit(X_train, y_train)
```

```python
y_pred1 = ridge.predict(X_test)
```

```python
mse = mean_squared_error(y_test,y_pred1)
print(mse)
mae = mean_absolute_error(y_test, y_pred1)
print(mae)
rmse = np.sqrt(mse)
print(rmse)
```

### **Lasso regression**

```python
from sklearn.linear_model import Lasso
lasso = Lasso()
```

```python
lasso = Lasso(alpha=20.0)
lasso.fit(X_train, y_train)
```

```python
y_pred2 = lasso.predict(X_test)
```

```python
mse = mean_squared_error(y_test,y_pred2)
print(mse)
mae = mean_absolute_error(y_test, y_pred2)
print(mae)
rmse = np.sqrt(mse)
print(rmse)
```

### **ElasticNet regression**

```python
from sklearn.linear_model import ElasticNet
elasticnet = ElasticNet()
```

```python
elasticnet = ElasticNet(alpha=20.0)
elasticnet.fit(X_train, y_train)
```

```python
y_pred3 = elasticnet.predict(X_test)
```

```python
mse = mean_squared_error(y_test,y_pred3)
print(mse)
mae = mean_absolute_error(y_test, y_pred3)
print(mae)
rmse = np.sqrt(mse)
print(rmse)
```
