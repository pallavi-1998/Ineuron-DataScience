# task - 1

#### to Split nested dictionary into separate columns in Pandas DataFrame

```python
import requests
url = "https://api.github.com/repos/pandas-dev/pandas/issues"
data = requests.get(url)
data1 = data.json()
df2 = pd.DataFrame(data1, columns= ['url', 'repository_url', 'labels_url', 'user'])
print(df2)
x = df2['user'].apply(pd.Series)#we first fetched column user as a Series, we then called apply(pd.Series), which returned a DataFrame where the column labels are the keys of the dictionaries.
df = pd.concat([df2, x], axis=1) #  to append the separated columns to df, use concat(~). Here, axis=1 means we are horizontally stacking the two DataFrames.
print(df)
#Finally, to remove column A now that you've unpacked its dictionaries:
df.drop(columns="user")
print(df)
#Here, a new DataFrame is returned and df is kept intact (still holds column user). To directly modify df, set inplace=True.
df.drop(columns='user', inplace=True)
print(df)
```
