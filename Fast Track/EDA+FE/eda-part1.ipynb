```python
features_with_nan = [features for features in dataset.columns if dataset[features].isnull().sum() >= 1]
```

```python
#features with nan values
features_with_nan
```

```python
dataset['Electrical'].isnull().mean()
```

```python
for feature in features_with_nan:
    print(feature, np.round(dataset[feature].isnull().mean()*100,4), '% missing values')
```

### let's have some changes if the feature has NAN values keep it as 1 and Non-NAN values will be as 0 
### this changes has been performed to find relationship between independent(other col on which basis dependent values are calculated) and dependent(col - SalesPrice) variable

```python
data = dataset.copy()
for feature in features_with_nan:
    # let's make a variables that indicates 1 if the observation was missing or 0 otherwise
    data[feature] = np.where(data[feature].isnull(),1,0)

    #let's calculate the mean SalePrice where the info is missing or present
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.title(feature)
    plt.show()
```

```python
sns.boxplot(dataset['SalePrice'])  # many outlires so median is used isabove program to get more precise result
```

#### Observation from this plot -->  Here with the relation between the missing values and the dependent varibles is clearly visible. so we need to replace these nan values with something meaningfull which we will do in the feature engineering section.

## numerical variables 

```python
## list of numerical variables

dataset['SalePrice'].dtype != 'O'  #'O' is same as 'object'
```

```python
numerical_features = [feature for feature in dataset.columns if dataset[feature].dtype != 'O']
```

```python
print(len(numerical_features))
dataset[numerical_features].head()
```

```python
dataset.info()
```

```python
## Temporal variables(Eg: Datetime variables)

Year_features = [feature for feature in numerical_features if 'Yr' in feature or 'Year' in feature]
```

```python
Year_features
```

```python
# finding all unique values from features 
for feature in Year_features:
    print(feature, dataset[feature].unique())
```

```python
dataset.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel("Year Sold")
plt.ylabel('Median House Price')
plt.title('House Price vs Year Sold')
```

```python
# Here we will compare the difference between all Year feature with Sale Price

data = dataset.copy()

for feature in Year_features:
    if feature != 'YrSold':
        data[feature] = data['YrSold'] - data[feature]
        plt.scatter(data[feature], data['SalePrice'])

        plt.xlabel(feature)
        plt.ylabel('Sale Price')
        plt.show()
```

### observation

#### observation from above 
Within all the yearwise info, it is clearly visible that the new houses are more costlier than the older one also grapg looks like pareto or power law distribution.

```python
### numerical variables- 2-types
#1. Continuous variable and Discrete variable

discrete_feature =  [feature for feature in numerical_features if len(dataset[feature].unique()) <= 25]
print(len(discrete_feature))
```

```python
dataset[discrete_feature].head()
```

```python
## let's find the relationship betweeen discrete and Sale Price

data = dataset.copy()
for feature in discrete_feature:
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('Sale Price')
    plt.title(feature)
    plt.show()
```

```python
## there iS a  relationship between Discrete feature and Sale Price
## continuous variable

continuous_feature = [feature for feature in numerical_features if feature not in discrete_feature+Year_features]
```

```python
print(len(continuous_feature))
```

```python
## let's analysis the continuous values by creating histogram to understand the distribution
data = dataset.copy()
for feature in continuous_feature:
    data.groupby(feature)['SalePrice'].median().plot.hist()
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.title(feature)
    plt.show()
```

## EDA part-2

```python
## we will be using logirithimic trasformation
data =  dataset.copy()
for feature in continuous_feature:
    if 0 in data[feature].unique():
        pass
    else:
        data[feature] = np.log(data[feature])
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('Sale Price')
        plt.title(feature)
        plt.show()
```

```python
### outliers
data = dataset.copy()
for feature in continuous_feature:
    if 0 in data[feature].unique():
        pass
    else:
        data[feature] = np.log(data[feature])
        data.boxplot(column=feature)
        plt.ylabel(feature)
        # plt.xlabel()
        plt.title(feature)
        plt.show()
```
