##### 1. take a data set from attached URL https://archive.ics.uci.edu/ml/datasets/glass+identification

```python
import mysql.connector as conn
import csv
```

```python
try:
    sqlCon = conn.connect(host = 'localhost', user = 'root', password = 'mysql')
except Exception as e:
    print("Connection failed ",e)
```

```python
sqlCon
```

```python
cursor = sqlCon.cursor()
```

```python
#create table for glass dataset
cursor.execute("CREATE DATABASE glass_db")
```

```python
#create table inside glass_db
try: 
    cursor.execute("use glass_db")
    query = "CREATE TABLE glass(id VARCHAR(8), refractive_index VARCHAR(8), Sodium VARCHAR(8), Magnesium VARCHAR(8), Aluminum VARCHAR(8), Silicon VARCHAR(8), Postassium VARCHAR(8), Calcium VARCHAR(8), Barium VARCHAR(8), Iron VARCHAR(8), Type_of_glass VARCHAR(8))"
    cursor.execute(query)
except Exception as e:
    print("Table creation failed ",e)
```

```python
cursor.execute("SHOW TABLES")
```

```python
cursor.fetchall()
```

```python
with open('glass.data','r') as file:
    glass_data = csv.reader(file, delimiter='\n')
    for i in glass_data:
        row_data = i[0].split(',')
        print(row_data)
```

```python
#dumping data
with open('glass.data','r') as file:
    glass_data = csv.reader(file, delimiter='\n')
    for i in glass_data:
        row_data = i[0].split(',')
        query = "INSERT INTO glass_db.glass VALUES('{}','{}','{}','{}','{}','{}','{}','{}','{}','{}','{}')".format(str(row_data[0]),str(row_data[1]),str(row_data[2]),str(row_data[3]),str(row_data[4]),str(row_data[5]),str(row_data[6]),str(row_data[7]),str(row_data[8]),str(row_data[9]),str(row_data[10]) )
        cursor.execute(query)
sqlCon.commit()
```

#### Alternate ways to dump data  

```python
try: 
    cursor.execute("use glass_db")
    query = "CREATE TABLE glass1(id VARCHAR(8), refractive_index VARCHAR(8), Sodium VARCHAR(8), Magnesium VARCHAR(8), Aluminum VARCHAR(8), Silicon VARCHAR(8), Postassium VARCHAR(8), Calcium VARCHAR(8), Barium VARCHAR(8), Iron VARCHAR(8), Type_of_glass VARCHAR(8))"
    cursor.execute(query)
except Exception as e:
    print("Table creation failed ",e)
```

```python
with open('glass.data','r') as file:
    glass_data = csv.reader(file, delimiter='\n')
    for i in glass_data:
        row_data = tuple(i[0].split(','))
        query = "INSERT INTO glass_db.glass1 VALUES {}".format(row_data)
        cursor.execute(query)
sqlCon.commit()
```

#### try to update and groupby
